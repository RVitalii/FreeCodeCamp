Complete the Category class in budget.py. It should be able to instantiate objects based on different budget categories like food, clothing, and entertainment. * A withdraw method that is similar to the deposit method, but the amount passed in should be stored in the ledger as a negative number. If there are not enough funds, nothing should be added to the ledger. This method should return True if the withdrawal took place, and False otherwise.
* A get_balance method that returns the current balance of the budget category based on the deposits and withdrawals that have occurred.
* A transfer method that accepts an amount and another budget category as arguments. The method should add a withdrawal with the amount and the description "Transfer to [Destination Budget Category]". The method should then add a deposit to the other budget category with the amount and the description "Transfer from [Source Budget Category]". If there are not enough funds, nothing should be added to either ledgers. This method should return True if the transfer took place, and False otherwise.
* A check_funds method that accepts an amount as an argument. It returns False if the amount is greater than the balance of the budget category and returns True otherwise. This method should be used by both the withdraw method and transfer method.

When the budget object is printed it should display:

The chart should show the percentage spent in each category passed in to the function. The percentage spent should be calculated only with withdrawals and not with deposits.
```python
#Creating a class
class Category():

    #appending attributes;
    def __init__(self, name):
        self.name = name
        self.expenses = 0
        self.income = 0
        self.ledger = []

    #calculating a total balance;
    def get_balance(self):
        return self.income + self.expenses

    #check if the amount is not exceeding the balance; 
    def check_funds(self, amount):
        if amount <= self.get_balance():
            return True
        else: 
            return False 
    
    #depositing money to a given account;
    def deposit(self, amount, desc = ""):
        self.ledger.append({"amount": amount, "description": desc})
        self.income = self.income + amount

    #withdrawing money from a given account; 
    def withdraw(self, amount, desc = ""):
        if self.check_funds(amount): 
            self.ledger.append({"amount": -amount, "description": desc})
            self.expenses = self.expenses - amount
            return True
        else:
            return False

    #transfering an amount of money from one given account to another one; 
    def transfer(self, amount, obj):
        if self.withdraw(amount, f"Transfer to {obj.name}"):
            obj.deposit(amount, f"Transfer from {self.name}")
            return True
        else:
            return False

    #altering the printing method of objects;
    def __str__(self):
        body = []
        
        header = self.name.center(30, '*')
        body.append(header)

        for pos in self.ledger:
            piece1 = pos.get("description")[:23]
            piece2 = f"{pos.get('amount'):.2f}"[:7]
            spaces = (30 - len(piece1) - len(piece2)) * ' '
            row = piece1 + spaces + piece2
            body.append(row)

        total = f"Total: {self.get_balance():.2f}"
        body.append(total)
        
        return '\n'.join(body)
```

```python
def create_spend_chart(categories):
    """
    DESC: The function creates and prints a barplot of expenses from a list Category objects. 
    INPUT: categories -- a list of objects from the Category class. 
    OUTPUT: a barplot of expenses by a category. 
    """
    
    #calculating the total expenses and the heights of perticular bars;
    total = sum([category.expenses for category in categories])
    bar_heights = {category.name: int(category.expenses / total * 10) for category in categories}
    
    #intiating a 'chart' variable and appending a title to it; 
    chart = []
    title = "Percentage spent by category"
    chart.append(title)

    #creating top part of the chart and appending it to the 'chart' variavble by one row at the time;
    for i in range(10, -1, -1):
        spaces = (3 - len(str(i * 10))) * " "
        level = f"{i * 10}| "
        row = spaces + level
        for category in categories:
            if bar_heights[category.name] >= i:
                row += "o  "
            else:
                row += 3*" "
        chart.append(row)
    
    #creating and appending the x-axis;
    chart.append(4*" " + (3*len(categories)+1)*"-")

    #creating and appending the bottom part to the main body; 
    max_len = max(list(map(len, bar_heights.keys())))
    for i in range(0, max_len):
        row = 5*" "
        for category in categories:
            if len(category.name)-1 >= i:
                row += category.name[i] + 2*" "
            else:
                row += 3*" "
        chart.append(row)

    #printing the chart;
    return('\n'.join(chart))
```

```python
#Example
food = Category("Food")
business = Category("Business")

food.deposit(900, "deposit")
business.deposit(900, "deposit")
food.withdraw(45.67, "groceries")
business.withdraw(200, "taxes")

print(food)
print('')
print(create_spend_chart([business, food]))
```
